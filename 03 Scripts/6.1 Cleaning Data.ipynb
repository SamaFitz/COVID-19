```python
# According to 'Footnote' column one or more data cells have counts
#    between 1-9 and have been suppressed in accordance with NCHS
#    confidentiality standards.
# For remaining death columns, will replace missing value with mean of 
#    1-9, which is 5.
```

```python
covid['COVID-19 Deaths'].fillna(5, inplace=True)
```
